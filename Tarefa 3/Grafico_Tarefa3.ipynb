```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler os dados do arquivo
dados = pd.read_csv('dados_grafico.txt', sep='\t', header=None, names=['Iteracoes', 'Digitos_Corretos', 'Tempo_Execucao'])

# Criando a figura e os subgráficos
fig, ax1 = plt.subplots()

# Plotando Digitos Corretos no eixo y
ax1.set_xlabel('Iterações')
ax1.set_ylabel('Dígitos Corretos', color='tab:blue')
ax1.plot(dados['Iteracoes'], dados['Digitos_Corretos'], color='tab:blue', label='Dígitos Corretos')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Criando um segundo eixo y para o tempo de execução
ax2 = ax1.twinx()
ax2.set_ylabel('Tempo de Execução (segundos)', color='tab:red')
ax2.plot(dados['Iteracoes'], dados['Tempo_Execucao'], color='tab:red', label='Tempo de Execução', linestyle='--')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Adicionando título
plt.title('Dígitos Corretos e Tempo de Execução vs Iterações')

# Exibindo a legenda
fig.tight_layout()
plt.show()

# Salvando o gráfico como imagem PNG
fig.savefig('grafico.png')
```
