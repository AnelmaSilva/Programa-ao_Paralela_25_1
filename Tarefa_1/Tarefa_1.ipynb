```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler os dados
data = pd.read_csv("tempos_execucao.csv")

# Plotar o grfico
plt.figure(figsize=(10, 6))
plt.plot(data["Tamanho"], data["Multiplicacao_1"], label="Acesso por Linhas", marker='o')
plt.plot(data["Tamanho"], data["Multiplicacao_2"], label="Acesso por Colunas", marker='x')

plt.xlabel("Tamanho da Matriz")
plt.ylabel("Tempo de Execuo (s)")
plt.title("Comparao de Tempo: Acesso por Linhas vs. Colunas")
plt.legend()
plt.grid(True)
plt.xscale("linear")  # Escala logartmica para melhor visualizao
plt.yscale("linear")  # Se os tempos variarem muito
plt.savefig("comparacao_tempos.png")
plt.show()
```
